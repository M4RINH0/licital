```python
import re
import pandas as pd

def encontra_cpf(texto):
    # Expressão regular para identificar um CPF no formato XXX.XXX.XXX-XX
    padrao_cpf = r'\b\d{3}\.\d{3}\.\d{3}-\d{2}\b'

    # Encontra todos os padrões de CPF na string
    cpfs_encontrados = re.findall(padrao_cpf, texto)

    return cpfs_encontrados


cpfs = encontra_cpf(text)

#dataframe
dic = {"cpfs": cpfs}
df0=pd.DataFrame(dic)
df0
```

```python

import re
def extract_person_names_regex(text):
    # Expressão regular para identificar nomes de pessoas (baseada em um padrão simples)
    pattern = re.compile(r'\b[A-Z][a-z]*\s[A-Z][a-z]*\b')

    # Encontra todos os padrões de nomes na string
    person_names = pattern.findall(text)

    return person_names



# Identifica nomes de pessoas usando expressões regulares
names_regex = extract_person_names_regex(text)
```

```python
import pandas as pd
#tira valores repetidos
names_regex_2= list(set(names_regex))
nomes = {"nomes": names_regex_2}
df = pd.DataFrame(nomes)

# Mostrar o DataFrame
df
```
